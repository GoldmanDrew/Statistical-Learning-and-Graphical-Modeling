## Bernoulli data and Beta prior
Suppose you are given a sample of $n$ iid Bernoulli random variables with probability of success equal to $\theta$. 

__Task:__ (5 pts) Write a function that produces the frozen posterior of $\theta$ given a Beta prior with pramaters $\alpha$ and $\beta$. Recall that a frozen RV is an object representing an RV with fixed parameters, e.g., 
```
rv = st.beta(a, b)
```
and can be called for example to compute the pdf at a certain point as 
```
rv.pdf(0.3)
```

````
def bayesBernBeta(data,prior):
    """ 
    input:
    - data is an iid sample of Bernoulli RVs
    - prior=[alpha,beta], prior Beta parameters
    """
    # Begin Solution
    a, b = prior
    a_posterior = a + len(data) * (sum(data)/len(data))
    b_posterior = b + len(data) - (len(data) * (sum(data)/len(data)))
    return st.beta(a_posterior,b_posterior)
    # End Solution
````

We will use three different beta priors:
- non-informative: uniform: $Beta(1,1)$.
- non-informative: Jeffery's: $Beta(1/2,1/2)$.
- informative: we believe that the prior peaks at $\theta = 1/3$ and furthermore, we believe that the probability that $\theta<0.5$ is $60\%$.

__Task:__ (15 pts) For the last case, we must find the $\alpha$ and $\beta$ parameters for a Beta prior that satisfies the condition. Generate 9 posterior pdf plots, for different data and prior combinations and discuss their differences.

````
# The priors
r = np.linspace(0,1,1000)
prior_unif = [1,1]
prior_jeff = [1/2,1/2]
# Finding the informative prior (last prior):
# we want the mode be at 1/3, so alpha = (beta+1)/2
# we numerically find the value of alpha using the cdf
alpha = np.linspace(0.51,10,1000) # we start from .51 since we need beta=2*alpha-1 > 0
cdf = np.array([st.beta.cdf(0.5,a,(2*a-1)) for a in alpha])
alpha_inf = alpha[sum(cdf<0.60)];
beta_inf = (2*alpha_inf-1)
prior_inf = [alpha_inf, beta_inf]
# Producing the plots

# Begin Solution

#unif
plt.plot(r, bayesBernBeta(small_sample, prior_unif).pdf(r), color = 'Green')
plt.title("Small N, Uniform Prior")
plt.show()
plt.plot(r, bayesBernBeta(med_sample, prior_unif).pdf(r), color = 'Red')
plt.title("Medium N, Uniform Prior")
plt.show()
plt.plot(r, bayesBernBeta(large_sample, prior_unif).pdf(r), color = 'Blue')
plt.title("Large N, Uniform Prior")
plt.show()

#jeff
plt.plot(r, bayesBernBeta(small_sample, prior_jeff).pdf(r), color = 'Green')
plt.title("Small N, Jeffrey's Prior")
plt.show()
plt.plot(r, bayesBernBeta(med_sample, prior_jeff).pdf(r), color = 'Red')
plt.title("Medium N, Jeffrey's Prior")
plt.show()
plt.plot(r, bayesBernBeta(large_sample, prior_jeff).pdf(r), color = 'Blue')
plt.title("Large N, Jeffrey's Prior")
plt.show()

#inf
plt.plot(r, bayesBernBeta(small_sample, prior_inf).pdf(r), color = 'Green')
plt.title("Small N, Informative Prior")
plt.show()
plt.plot(r, bayesBernBeta(med_sample, prior_inf).pdf(r), color = 'Red')
plt.title("Medium N, Informative Prior")
plt.show()
plt.plot(r, bayesBernBeta(large_sample, prior_inf).pdf(r), color = 'Blue')
plt.title("Large N, Informative Prior")
plt.show()

print("It is clear for all three priors that as the sample size increases, the effect that the prior has on the posterior decreases.") 
print("However, for the uniform prior, the distributions are clearly more spread out with a larger variance.")
print("Whereas, the jeff prior is similar to the uniform prior, but more of the probability density lies further right on the distribution (right skewed).")
print("Finally, the informative prior has more density distribution aroudn 1/3, and less around everything else, for the smaller sample sizes, of course.")
# End Solution
````
